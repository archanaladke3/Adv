##Dictonary
data stored in key value pair

```python
student= {
    "name":"priya",
    "age":29,
    "sub":"math",

}
print(student)
print(student["name"])
```

```python
#updating dictonary
student["school"]="vidhyamandir"
print(student)
```

```python
student["rollno"]=121
```

```python
# udpate age 29 to 11
student["age"]=11
print(student)
```

```python
# create dic with empcode,name,salary
empinfo = {
    "empcode":121,
         "empname":"jolly",
          "salary":30000
    }

print(empinfo)
```

```python
#add dept, country city

empinfo["department"]="IT"
empinfo["city"]="mumbai"
empinfo["country"]="india"
print(empinfo)
```

```python
###Dictonary method
keys()method:
keys() method returns the all keys present in dectonary
```

```python
empinfo.keys()
```

value()method
values() method returns all the method stored in dictonary

```python
empinfo.values()
```

Items()
in items() methods return both keys and value present in dictonary

```python
empinfo.items()
```
