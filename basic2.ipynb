```python
#updating dictonary
student["school"]="vidhyamandir"
print(student)
```

```python
student["rollno"]=121
```

```python
# udpate age 29 to 11
student["age"]=11
print(student)
```

```python
# create dic with empcode,name,salary
empinfo = {
    "empcode":121,
         "empname":"jolly",
          "salary":30000
    }

print(empinfo)
```

```python
#add dept, country city

empinfo["department"]="IT"
empinfo["city"]="mumbai"
empinfo["country"]="india"
print(empinfo)
```
